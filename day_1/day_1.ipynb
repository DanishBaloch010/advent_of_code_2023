```python
file_path = 'data.txt'  # Replace 'your_file.txt' with the path to your text file

try:
    counter = 0
    add = 0

    with open(file_path, 'r') as file:
        for line in file:            
            text = line.strip()
            
            print(text)
            
            first_num = ''
            second_num = ''
            for idx , val in enumerate(text):
                if text[idx].isdigit():
                    first_num = str(text[idx])
                    break
            
            for idx , val in enumerate(text):
                if text[-(idx+1)].isdigit():
                    second_num = str(text[-(idx+1)])
                    break
            
            stitch = int(first_num + second_num)
            
            add = add + stitch
                    
            counter = counter + 1

    print(counter)
    print(add)
except FileNotFoundError:
    print(f"The file '{file_path}' was not found.")
except Exception as e:
    print(f"An error occurred: {e}")
```

```python
with open("data.txt") as fin:
    data = fin.read()

ans = 0
nums = ["one", "two", "three", "four", "five",
        "six", "seven", "eight", "nine"]

for line in data.strip().split():
    first = None
    last = None
    for i in range(len(line)):
        cur = None
        c = line[i]
        if c.isdigit():
            cur = int(c)
        for j, num in enumerate(nums):
            if line[i:(i+len(num))] == num:
                cur = j + 1
                break 
        if cur:
            if first == None:
                first = cur

            last = cur
    ans += first * 10 + last
print(ans)
```
